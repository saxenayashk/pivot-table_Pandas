```python
#import the libraries
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
#Load the dataset
data=sns.load_dataset('tips')
data.head(10)
```

```python
#To display the average total bill per gender
data.groupby('sex')[['total_bill']].mean()
```

```python
#To display the average total bill per gender and smoking
data.groupby(['sex','smoker'])[['total_bill']].mean().unstack()
```

```python
#pivot table
data.pivot_table(index='sex',columns='smoker',values='total_bill')
```

```python
#aggfunc
data.pivot_table(index='sex',columns='smoker',values='total_bill',aggfunc='sum')
```

```python
data.pivot_table(index='sex',columns='smoker',values='total_bill',aggfunc='count')
```

```python
data.pivot_table(index='sex',columns='smoker',values='total_bill',aggfunc='std')
```

```python
#Multidimensional pivot-table
data.pivot_table(index=['sex','smoker'],columns=['day','time'],values='total_bill')
```

```python
data.pivot_table(index='sex',columns='smoker',values='total_bill',aggfunc='sum',margins=True)
```
